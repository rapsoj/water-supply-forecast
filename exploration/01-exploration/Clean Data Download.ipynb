```python
import requests as req
import pandas as pd

url = 'https://raw.githubusercontent.com/rapsoj/water-supply-forecast/main/assets/data/teleconnections/mjo.txt'
res = req.get(url)

file = open('mjo.txt', 'w')
file.write(res.text)
file.close()
```

```python
filename = 'oni.txt'

url = 'https://raw.githubusercontent.com/rapsoj/water-supply-forecast/main/assets/data/teleconnections/' + filename
res = req.get(url)
file = open(filename, 'w')
file.write(res.text)
file.close()

df = pd.read_csv(filename, delim_whitespace = True)
df.to_csv('oni.csv')
```
