```python
import requests
import csv
```

```python
api_url = "https://data.rcc-acis.org/StnMeta"

states = ['MT', 'ID', 'CO', 'WY', 'WA', 'UT', 'CA', 'OR', 'NM']
format_param = 'json'

try:
    all_data = []

    # Iterate over each state and make a request
    for state in states:
        params = {
            'state': state,
            'format': format_param,
        }

        response = requests.get(api_url, params=params)
        response.raise_for_status()

        state_data = response.json().get('meta', [])
        all_data.extend(state_data)

    csv_file = "acis_stnmeta.csv"

    # Extract all unique fields from the data
    all_fields = set()
    for station_info in all_data:
        all_fields.update(station_info.keys())

    # Open the CSV file in write mode
    with open(csv_file, 'w', newline='') as csvfile:
        # Create a CSV writer object
        csvwriter = csv.writer(csvfile)

        # Write the header (field names) to the CSV file
        csvwriter.writerow(all_fields)

        # Write each row of data to the CSV file
        for station_info in all_data:
            # Extract all values dynamically based on the available fields
            row = [station_info.get(field, "") for field in all_fields]
            csvwriter.writerow(row)

    print(f"Data saved to {csv_file}")
except requests.exceptions.HTTPError as errh:
    print(f"HTTP Error: {errh}")
except requests.exceptions.RequestException as err:
    print(f"Request Error: {err}")
```
