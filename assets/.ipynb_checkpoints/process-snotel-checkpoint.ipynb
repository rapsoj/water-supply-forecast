# Process SNOTEL

### Prepare Workspace

```python
# Import system libraries
import os
import sys

# Import data manipulation libraries
import pandas as pd
import numpy as np
import datetime

# Import statistical libraries
import statsmodels.api as sm

# Set working directory
os.chdir('/Users/jessicarapson/Documents/GitHub/water-supply-forecast')
```

### Download Data from Files

```python
# Download snowpack and streamflow data
df_snotel = pd.read_csv('assets/data/snotel/snotel.csv')
df_cdec = pd.read_csv('assets/data/cdec/cdec.csv')
df_train = pd.read_csv('assets/data/train.csv')

# Link CDEC and SNOTEL datas
snow_dict = {'SNOW WC': 'WTEQ_DAILY', 'TEMP AV': 'TAVG_DAILY', 'TEMP MX': 'TMAX_DAILY',
             'TEMP MN': 'TMIN_DAILY', 'RAIN': 'PREC_DAILY','SNOW DP': 'SNWD_DAILY',
             'station_id': 'station', 'site_id': 'site'}
df_cdec = df_cdec.rename(columns=snow_dict)[df_snotel.columns]
df_snowpack = pd.concat([df_snotel, df_cdec], axis=0)

# Convert 'date_column' to datetime format
df_snowpack['date'] = pd.to_datetime(df_snowpack['date'])

# Function to calculate water year
def get_water_year(date):
    if date.month >= 10:  # If month is October or later, consider it part of the next water year
        return date.year + 1
    else:
        return date.year

# Link data on water year
df_snowpack['water_year'] = df_snowpack['date'].apply(get_water_year)
df_combined = pd.merge(df_snowpack, df_train, left_on=['site','water_year'],
                       right_on=['site_id','year'], how='left')
df_combined.drop(['site_id', 'year'], axis=1, inplace=True)

# Filter to training years
df_combined = df_combined[df_combined['water_year'] >= 1985]
```

### Get Correlation-Weighted Averages

```python
# Extract value from April 1st
data = df_combined[(df_combined['date'].dt.month == 4) & (df_combined['date'].dt.day == 1)]

# Intialise list of correlations
corr_list = []

# Loop through sites
sites = df_combined.site.unique()
for i in sites:

    # Subset data to site of interest
    cols = ['station', 'SNWD_DAILY', 'volume', 'date']
    data_site = data[data['site'] == i][cols]

    # Pivot data
    data_pivot = data_site.pivot_table(
        index=['date', 'volume'], columns='station', values='SNWD_DAILY').reset_index()

    # Initialise dataframe to store data
    stations = data_pivot.columns[2:]
    corrs = pd.DataFrame({'site': i, 'station': stations, 'sample_size': np.nan,
                          'coeff': np.nan, 'r2': np.nan, 'p': np.nan})

    # Loop through stations
    for j in stations:

        # Subset to station of interest
        data_station = data_pivot[['volume', j]]

        # Drop rows with missing data
        data_station = data_station.dropna(axis=0)
        corrs.loc[corrs['station'] == j, 'sample_size'] = len(data_station)

        # Select data for linear regression
        X = sm.add_constant(data_station.drop('volume', axis=1))
        y = data_station['volume']

        # Fit a linear regression model
        model = sm.OLS(y, X).fit()

        # Extract values of interest
        corrs.loc[corrs['station'] == j, 'coeff'] = model.params.iloc[1]
        corrs.loc[corrs['station'] == j, 'r2'] = float(model.rsquared)
        corrs.loc[corrs['station'] == j, 'p'] = float(model.pvalues.iloc[1])

    corr_list.append(corrs)
    
# Combine data
corr_full = pd.concat(corr_list)
```
