```python
import os
import pandas as pd
from consts import ORDERED_SITE_IDS, FIRST_FULL_GT_YEAR
import matplotlib.pyplot as plt
path = os.getcwd()
```

```python
calculated_labels = pd.read_csv(os.path.join(path, '..', 'assets', 'data', 'additional_train.csv'))
old_labels = pd.read_csv(os.path.join(path, '..', 'assets', 'data', 'train.csv'))
old_labels = old_labels[old_labels.year >= FIRST_FULL_GT_YEAR]
```

```python
additional_site_ids = calculated_labels.site_id.unique()
calculated_labels_for_old_sites = calculated_labels[calculated_labels.site_id.isin(ORDERED_SITE_IDS)]
plt.hist([calculated_labels_for_old_sites.volume, old_labels.volume])
plt.title('Distribution of newly calculated labels and old labels for old sites')
print(f'Mean of calculated labels for old sites: {calculated_labels_for_old_sites.volume.mean()}')
print(f'Mean of old labels for old sites: {old_labels.volume.mean()}')
print(f'Num of old sites present in new labels: {calculated_labels_for_old_sites.site_id.nunique()}')
plt.show()
```

For individual sites, how are the calculated labels looking?

```python
for site_id in ORDERED_SITE_IDS:
    calc_mask = calculated_labels_for_old_sites.site_id==site_id
    old_mask = old_labels.site_id==site_id
    plt.scatter(calculated_labels_for_old_sites.year[calc_mask], calculated_labels_for_old_sites.volume[calc_mask])
    plt.scatter(old_labels.year[old_mask], old_labels.volume[old_mask])
    plt.title(site_id)
    plt.show()
```

Compare final distributions of new calculated labels and old


```python
plt.hist([calculated_labels.volume, old_labels.volume], bins=50, density=True)
plt.title('Distribution of newly calculated labels (all sites) and old')
print(f'Mean of calculated labels for all sites: {calculated_labels.volume.mean()}')
print(f'Mean of old labels for old sites: {old_labels.volume.mean()}')
print(f'Num of old sites present in new labels: {calculated_labels.site_id.nunique()}')
plt.show()
```
